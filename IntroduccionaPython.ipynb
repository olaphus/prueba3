## 1. Títulos, tipos de fuentes y tablas

A medida que se ha escrito este texto, ya he utilizado dos aspectos básicos de **Jupyter Notebook**. Note que con los asteriscos (\*) y otros símbolos, se puede volver un sector del texto en cursiva, en negrilla, o en ambos, etc., de acuerdo a la siguiente tabla:

|Texto en Markdown|Como aparece compilado|
|:---|:---:|
|\*Cursiva\*|*Cursiva*|
|\*\*Negrilla\*\*|**Negrilla**|
|\*\*\*Cursiva y negrilla\*\*\*|***Cursiva y negrilla***|
|\`Fuente tipo código\`|`Fuente tipo código`|
|\~ Tachado \~|~Tachado~|
|\_Otra Cursiva\_|_Otra cursiva_|

Fíjese que para que los caracterese especiales de Markdown aparezcan compilados, se necesita poner antes el símbolo: \\. Note que a lo largo del texto también se ha usado el símbolo numeral (\#) para comenzar títulos y subtítulos.

## 2. Introducción a los *DataFrames*

Si recordamos nuestra clase de R. Acá importaremos unos datos que nos son familiares. En la siguiente línea se importa la base de datos de las flores del género *Iris*. Note que la especie es dada en forma de número.

````python
!pip install sklearn
````

````python
from sklearn import datasets
iris =datasets.load_iris()
df=pd.DataFrame(iris.data)
df.columns=iris.feature_names
df['Especie']=iris.target
df.head(10)
````

Los nombres de las especies son los siguientes.

````python
specnames=iris.target_names
specnames
````

Podemos ahora modificar la columna de especie para cambiar los números por los nombres de las especies.

````python
df.Especie=df.Especie.apply(lambda x: specnames[x])
df.head(7)
````

Exploremos un poco los datos

````python
df.info()
````

Podemos generar un histograma rápido de esa variable.

````python
# No se olvide de correr esta línea para que se vean las gráficas
%matplotlib inline
ax=sns.distplot(df["sepal length (cm)"])

````

Exploremos una variable cuantitativa:

````python
print(df.Especie.value_counts())
sns.countplot(df.Especie)
````

Veamos tres histogramas a la vez.

````python
prop="sepal length (cm)"
subb1=df[df.Especie==iris.target_names[0]][prop]
subb2=df[df.Especie==iris.target_names[1]][prop]
subb3=df[df.Especie==iris.target_names[2]][prop]
ax=sns.distplot(subb1);ax=sns.distplot(subb2);ax=sns.distplot(subb3)
ax=plt.legend(iris.target_names)
ax=plt.xlabel("Longitud de sépalo")
````

Ahora exploremos la relación de dos variables cuantitativas.

````python
ax = sns.scatterplot(x="sepal length (cm)", y="sepal width (cm)", hue="Especie",
...                      data=df)
````
